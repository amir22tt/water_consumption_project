**model train**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.preprocessing import LabelEncoder
```

```python
import pandas as pd
df = pd.read_csv('process_data.csv')
```

```python
le = LabelEncoder()
df['Gender_Code'] = le.fit_transform(df['Gender'])
df['City_Code'] = le.fit_transform(df['City'])
```

```python
activity_map = {'Low': 0, 'Medium': 1, 'High': 2}
df['Activity_Code'] = df['Activity_Level'].map(activity_map)
```

```python
y = df['Water_Consumed_Liters']
```

```python
X = df[['Age', 'Temperature_C', 'Activity_Code', 'Gender_Code', 'City_Code']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=94)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
score = r2_score(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
```

```python
print(f"Score Accuracy fit: {score * 100:.2f}%")
print(f"Mean Absolute Error: {mae:.2f} Liters")
```

```python
import pickle
with open("model.pkl", "wb") as f:pickle.dump(model,f)
```
